## Exercise 1: Rewrite and Optimize a Vague Prompt
**Scenario**: Manager asks for a LinkedIn post about FlowNest with the vague prompt `"Write something about productivity tips."`

### Tasks
- Identify 3 critical issues with the prompt.
- Rewrite it using: role prompting, clear instructions, and format/tone/length constraints.
- Add a final line to make it reusable for future social posts.

```python
critical_issues = [
    'Lack of Role: It does not specify a persona (e.g., social media manager).',
    'Ambiguous Scope: "Productivity tips" is too broad; it needs a specific focus or brand like FlowNest.',
    'Missing Constraints: There are no instructions regarding tone, length, or target audience (LinkedIn professionals).'
]
critical_issues
```

```python
rewritten_prompt = """Act as a Senior Social Media Manager for FlowNest.
Write a professional yet engaging LinkedIn post (max 150 words) sharing 3 actionable productivity tips for remote teams.
Use an encouraging tone and include a call-to-action to visit flownest.com.

[Optional]: Replace 'remote teams' with [TARGET_AUDIENCE] for future use."""
rewritten_prompt
```

## Exercise 2: Multi-Part Prompt for Quiz Generation
**Scenario**: Create an educational prompt for 7th graders about volcanic eruptions.

### Tasks
- Ask for: 2-bullet summary + 3 MCQs (1 correct, 2 distractors).
- Require friendly tone, simple vocab (11?13 y/o), and Google Slides-ready output.
- Explain 3 ways your prompt improves on: `Make a quiz for kids about this article.`

```python
quiz_prompt = """You are an expert middle-school science teacher.
Summarize the provided article on volcanic eruptions in 2 simple bullet points for 12-year-olds.
Then, create 3 multiple-choice questions. Each question must have 1 correct answer and 2 plausible distractors.
Use a friendly tone and format the output clearly for a Google Slide layout."""
quiz_prompt
```

```python
improvements = [
    'Persona: Setting the role of a teacher ensures appropriate vocabulary levels.',
    'Structure: Explicitly asking for a summary and MCQs prevents the model from generating random facts.',
    'Audience Targeting: Specifying "12-year-olds" prevents the output from being too technical or too babyish.'
]
improvements
```

## Exercise 6: Hallucination Spotting and Mitigation
**Scenario**: Model added an unsupported claim in a climate/marine article summary.

### Tasks
1) Prompt the model to avoid hallucinations using only provided text.
2) Rewrite with verification language.
3) List 2 mitigation strategies.
4) Name 2 domains where hallucinations are risky and why.

```python
no_hallucination_prompt = """Summarize the provided text about marine life.
Only use information explicitly stated in the text.
If the text does not contain a specific answer, state 'Information not found.'
Do not include external knowledge."""
no_hallucination_prompt
```

```python
verification_prompt = """Summarize the article. For every claim you make, cite the specific sentence or paragraph from the source.
If you cannot find a direct citation, do not include the claim."""
verification_prompt
```

```python
mitigation_strategies = [
    'Grounding: Forcing the model to use only provided context.',
    'Chain of Verification: Asking the model to double-check its own facts before final output.'
]
risky_domains = {
    'domain_1': {'name': 'Medical', 'why': 'Incorrect dosage or diagnosis advice can lead to physical harm.'},
    'domain_2': {'name': 'Legal', 'why': 'Citing fake precedents can result in legal penalties and loss of case credibility.'},
}
print(mitigation_strategies)
print(risky_domains)
```
