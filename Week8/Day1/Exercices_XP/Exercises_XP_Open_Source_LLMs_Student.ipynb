## Exercise 1: Open Source Levels Reflection
**Use:** Fully Open, Weights Released, Architecture Only; model components; fine-tuning/domain adaptation; healthcare compliance.

**Step-by-step**
1) Gather definitions for the three openness levels.  
2) Identify key characteristics (what is open; what you can/cannot do).  
3) Compare side by side in a table (what's open vs. impact).  
4) Write a 3?5 sentence comparative paragraph.  
5) Answer the healthcare prompt (1?2 sentences) about retraining on clinical data.

**Deliverables:** paragraph + healthcare answer.

```python
open_source_levels = {
    "Fully Open": {
        "definition": "Models where the training data, code, weights, and documentation are all public.",
        "what_is_open": ["Training datasets", "Pre-processing code"],
        "what_you_can_do": ["Audit bias", "Fully reproduce the model"],
        "what_you_cannot_do": ["Claim exclusive ownership"],
    },
    "Weights Released": {
        "definition": "Models where only the final parameters (weights) are shared, often under a custom license.",
        "what_is_open": ["Model weights", "Inference code"],
        "what_you_can_do": ["Local deployment", "Fine-tuning"],
        "what_you_cannot_do": ["See training data", "Understand exact data filtering"],
    },
    "Architecture Only": {
        "definition": "Only the technical structure is described, but weights are kept private or behind an API.",
        "what_is_open": ["Technical papers", "Layer specifications"],
        "what_you_can_do": ["Re-implement structure", "Research design patterns"],
        "what_you_cannot_do": ["Run the model without training from scratch", "Access the original intelligence"],
    },
}
open_source_levels
```

```python
comparison_table = """| Openness level | What's open? | Impact on retraining/modifying |
| --- | --- | --- |
| Fully Open | Data, code, weights | Maximum; allows deep domain adaptation and auditing. |
| Weights Released | Final parameters | High; allows fine-tuning but data remains a black box. |
| Architecture Only | Structural design | Low; requires massive resources to train from zero. |"""
print(comparison_table)
```

```python
comparative_paragraph = """Fully open models provide the highest level of transparency, enabling developers to inspect the raw data for biases. In contrast, 'Weights Released' models allow for immediate deployment and fine-tuning but hide the underlying data distribution. 'Architecture Only' models are the most restrictive, offering only a blueprint without the actual learned intelligence. Choosing between them depends on the need for auditability versus the convenience of pre-trained performance."""
print(comparative_paragraph)
```

```python
healthcare_prompt_answer = """To retrain a model on sensitive clinical data, a 'Fully Open' or 'Weights Released' model is required to ensure data privacy by running the training locally on private infrastructure. Architecture-only models are unsuitable as they lack the baseline knowledge needed for efficient medical domain adaptation."""
print(healthcare_prompt_answer)
```
