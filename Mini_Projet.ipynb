```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Préparation de données d'exemple (MNIST)
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train)).batch(32)
test_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test)).batch(32)

# 2. Définition du modèle
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 3. Compilation du modèle
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
EPOCHS = 2

history = model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=EPOCHS
)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Redéfinition et entraînement rapide si le modèle est manquant
if 'model' not in locals():
    print("Initialisation du modèle et des données...")
    (x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
    x_train, x_test = x_train / 255.0, x_test / 255.0
    train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train)).batch(32)
    test_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test)).batch(32)

    model = models.Sequential([
        layers.Flatten(input_shape=(28, 28)),
        layers.Dense(128, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    model.fit(train_ds, epochs=1, verbose=1)

# Évaluation finale
eval_metrics = model.evaluate(test_ds)
print("\nTest Loss     :", eval_metrics[0])
print("Test Accuracy :", eval_metrics[1])
```

```python
from transformers import pipeline

# Initialisation du pipeline de sentiment analysis
# Cela télécharge et configure automatiquement le modèle et le tokenizer appropriés
sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

def predict_sentiment(text: str):
    # Le pipeline retourne une liste de dictionnaires
    result = sentiment_pipeline(text)[0]
    label = result['label']
    confidence = result['score']

    # Conversion du label pour correspondre au format attendu (SST-2 est souvent déjà POSITIVE/NEGATIVE)
    label = label.capitalize()
    return label, float(confidence)
```

2. Where would you add guardrails before deployment?

Answer:

Several guardrails should be added before deploying the model:

* Human review for low-confidence predictions.
* Monitoring for data drift over time.
* Bias testing across different customer groups.
* Escalation rules for highly negative messages.
* Logging and auditing of predictions for compliance and debugging.
* Periodic retraining using newly collected customer feedback.
