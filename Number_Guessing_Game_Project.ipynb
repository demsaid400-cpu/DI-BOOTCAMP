```python
# Étape 1 : Importation du module random
import random

# Étape 2 : Écriture de la logique du jeu
def number_guessing_game():
    # 2. Génération du nombre aléatoire entre 1 et 100
    random_number = random.randint(1, 100)

    # 3. Définition du nombre maximum de tentatives
    max_attempts = 7

    print("Bienvenue au jeu de devinette de nombre !")
    print(f"J'ai choisi un nombre entre 1 et 100. Vous avez {max_attempts} tentatives pour le trouver.\n")

    # Variable pour suivre si le joueur a gagné
    has_won = False

    # 4. Boucle for pour limiter les tentatives
    for attempt in range(1, max_attempts + 1):
        # 5. Demande de saisie à l'utilisateur et conversion en entier
        try:
            guess = int(input(f"Tentative {attempt}/{max_attempts} - Entrez votre choix : "))
        except ValueError:
            print("Veuillez entrer un nombre entier valide.")
            continue

        # 6. Comparaison avec les instructions if-elif-else
        if guess < random_number:
            print("Trop bas ! (Too low!)")
        elif guess > random_number:
            print("Trop haut ! (Too high!)")
        else:
            print(f"Félicitations ! Vous avez trouvé le bon nombre en {attempt} tentatives !")
            has_won = True
            break  # Fin du jeu si la réponse est correcte

    # 7. Message si le joueur a épuisé ses tentatives sans trouver
    if not has_won:
        print(f"\nDommage ! Vous avez épuisé toutes vos tentatives. Le nombre correct était : {random_number}")

# Appel de la fonction pour lancer le jeu
number_guessing_game()
```
