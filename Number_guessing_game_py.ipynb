```python
# Step 1: Set up the project
# Import standard pour la génération de nombres aléatoires
import random

# Step 2: Write the game logic
# 1. Create a function called number_guessing_game()
def number_guessing_game():
    print("Welcome to the Number Guessing Game!")
    print("I have picked a number between 1 and 100. Can you guess it?")

    # 2. Generate a random number between 1 and 100
    random_number = random.randint(1, 100)

    # 3. Define the maximum number of attempts
    max_attempts = 7

    # Variable pour suivre si le joueur a trouvé le nombre
    guessed_correctly = False

    # 4. Use a for loop to give the player a limited number of attempts
    for attempt in range(1, max_attempts + 1):
        # 5. Prompt the player to enter their guess and convert to integer
        guess = int(input(f"Attempt {attempt}/{max_attempts} - Enter your guess: "))

        # 6. Compare the player's guess with the random number
        if guess < random_number:
            print("Too low!")
        elif guess > random_number:
            print("Too high!")
        else:
            print(f"Congratulations! You guessed the correct number in {attempt} attempts!")
            guessed_correctly = True
            break # End the game early since they won

    # 7. If the player doesn't guess correctly within the allowed attempts
    if not guessed_correctly:
        print(f"Game Over! You've used all your attempts. The correct number was {random_number}.")

# Execution de la fonction
number_guessing_game()
```

### Explication du code : Le jeu de devinettes

Voici les concepts clés utilisés dans votre script :

1.  **`import random`** : On importe une bibliothèque qui permet de générer des éléments aléatoires.
2.  **`random.randint(1, 100)`** : Cette fonction choisit un nombre entier au hasard entre 1 et 100.
3.  **`input()`** : Cette fonction interrompt le programme et attend que l'utilisateur tape quelque chose au clavier. Le `int()` autour sert à transformer le texte tapé en un nombre entier.
4.  **La boucle `for`** : `for attempt in range(1, max_attempts + 1):` permet de répéter le bloc de code (demander un nombre) exactement 7 fois.
5.  **Les conditions `if / elif / else`** :
    *   `if guess < random_number` : Vérifie si la proposition est trop petite.
    *   `elif guess > random_number` : Vérifie si elle est trop grande.
    *   `else` : Si ce n'est ni plus petit, ni plus grand, c'est que le joueur a trouvé !
6.  **`break`** : Ce mot-clé permet de sortir immédiatement de la boucle `for` dès que le joueur a gagné, sans attendre d'arriver au 7ème essai.
