## 🌟 Exercise 2: LLM Architecture and Application Scenarios
For each, describe (a) core architectural differences, (b) a real application, (c) why it fits.

### BERT
- **Architecture**: Bidirectional encoder-only; uses Masked Language Modeling (MLM).
- **Application**: Search Engine Query Understanding (Google Search).
- **Why**: BERT's bidirectional nature allows it to understand the full context of a word by looking at both its left and right neighbors, making it excellent for understanding the intent behind complex queries.

### GPT
- **Architecture**: Decoder-only; autoregressive causal LM predicting the next token.
- **Application**: Creative Writing Assistants (ChatGPT).
- **Why**: The autoregressive design is optimized for generating coherent, fluent text by conditioning each new word on the sequence of all previous words.

### T5
- **Architecture**: Encoder-decoder; text-to-text framework treating every task as a string conversion.
- **Application**: Document Summarization.
- **Why**: The encoder-decoder structure is ideal for mapping an input sequence (long text) to a target sequence (summary) while maintaining a mapping between source and target.

## 🌟 Exercise 3: Benefits and Ethics of Pre-training
- **Benefits**:
  1) **Improved generalization**: Models learn universal language features that apply across many tasks.
  2) **Less labeled data**: High performance can be achieved with only a few hundred labeled examples during fine-tuning.
  3) **Faster fine-tuning**: Starting from a pre-trained state requires significantly fewer training epochs.
  4) **Transfer learning**: Knowledge from a large general corpus is 'transferred' to a specialized domain.
  5) **Robustness**: Exposure to vast amounts of text makes the model more resilient to variations in input.

- **Ethical concerns**: Models can amplify social biases present in training data, generate misinformation (hallucinations), be misused for phishing/propaganda, and raise privacy issues if training data contains PII.
- **Mitigations**: Using curated datasets, implementing differential privacy, applying safety filters, using RLHF to align with human values, and performing third-party audits.

## 🌟 Exercise 4: Transformer Architecture Deep Dive
### Self-Attention & Multi-Head Attention
- **Q/K/V Flow**: Queries (Q) represent what a token is looking for, Keys (K) represent what a token offers, and Values (V) contain the actual information. Softmax weighting of the Q-K dot products determines the attention scores used to mix the Values.
- **Multi-Head**: Multiple heads allow the model to attend to information in different subspaces simultaneously, capturing diverse relationships like syntax and semantics.
- **Example**: "The bank was closed because it was a holiday."
  - **Head 1**: Links "it" to "bank" (coreference resolution).
  - **Head 2**: Links "closed" to "holiday" (causal relationship).

### Pre-training Objectives
- **MLM vs CLM**: MLM (Masked) hides random tokens to learn bidirectional context; CLM (Causal) hides future tokens to learn generation.
- **Preference**: Use MLM for understanding/extraction; use CLM for generation.
- **NSP**: Next Sentence Prediction helped BERT understand relationships between sentences but was found unnecessary if the model is trained on longer contiguous sequences.

### Transformer Model Selection
- **Sentiment on reviews**: **Encoder-only (BERT)**; Justification: It excels at classification by understanding the entire context of the review.
- **Conversational chatbot**: **Decoder-only (GPT)**; Justification: Optimized for high-quality, creative sequence generation.
- **Technical translation**: **Encoder-Decoder (T5/BART)**; Justification: Naturally handles the mapping from one language sequence to another.

### Positional Encoding
- **Purpose**: Transformers process all tokens in parallel; positions provide the order information lost without recurrence.
- **Failure**: Without it, "Dog bites man" and "Man bites dog" would have identical representations.

## 🌟 Exercise 6: Softmax Temperature: The Randomness Regulator
### 1) Temperature Scenarios
- **T=0.2**: Highly focused and deterministic; chooses the most likely token (low entropy).
- **T=1.5**: Highly random and diverse; flattens the distribution (high entropy).
- **T=1.0**: Standard distribution; balances coherence and variety.

### 2) Application Design
- **Bedtime stories**: T=0.8 to 1.0. Encourages creativity and narrative variety while remaining coherent enough for a child to follow.
- **Financial report summaries**: T=0.1 or 0.2. Prioritizes factual accuracy and consistency over creative phrasing.

### 3) Temperature & Bias
- **Effect**: High temperature can surface fringe or biased associations by giving weight to unlikely tokens, whereas low temperature often reinforces the dominant biases found in the training data.
- **Example**: In an occupational prompt, low T might always produce "He" for "Doctor," while high T might introduce more variety but also more nonsensical or offensive pairings.

### (Optional) Quick Generation Demo Across Temperatures
Note: Requires `transformers` and model download; skip if offline.


```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_name = "gpt2"  # TODO: choose a causal LM
prompt = "Artificial intelligence will"  # TODO: write your own prompt

pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(model_name),
    tokenizer=AutoTokenizer.from_pretrained(model_name),
    device=0 if torch.cuda.is_available() else -1,
)

for T in [0.2, 1.0, 1.5]:
    out = pipe(prompt, max_new_tokens=40, temperature=T, do_sample=True)
    print("--- Temperature:", T)
    print(out[0]["generated_text"])  # Inspect how style changes with T
```
