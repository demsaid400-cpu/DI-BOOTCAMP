```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
```

```python
# Flatten
x_train_dense = x_train.reshape(x_train.shape[0], 28*28)
x_test_dense = x_test.reshape(x_test.shape[0], 28*28)

# Normalisation
x_train_dense = x_train_dense / 255.0
x_test_dense = x_test_dense / 255.0

# One-hot encoding
y_train_cat = keras.utils.to_categorical(y_train, 10)
y_test_cat = keras.utils.to_categorical(y_test, 10)
```

```python
dense_model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
dense_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_dense = dense_model.fit(
    x_train_dense, y_train_cat,
    epochs=5,
    batch_size=32,
    validation_split=0.2
)
```

```python
dense_loss, dense_acc = dense_model.evaluate(x_test_dense, y_test_cat)
print("Dense Test Accuracy:", dense_acc)
```

```python
x_train_cnn = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test_cnn = x_test.reshape(-1, 28, 28, 1) / 255.0
```

```python
cnn_model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),

    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
cnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_cnn = cnn_model.fit(
    x_train_cnn, y_train_cat,
    epochs=5,
    batch_size=32,
    validation_split=0.2
)
```

```python
cnn_loss, cnn_acc = cnn_model.evaluate(x_test_cnn, y_test_cat)
print("CNN Test Accuracy:", cnn_acc)
```

```python
print("Dense Accuracy:", dense_acc)
print("CNN Accuracy:", cnn_acc)
```
