```python
from collections import deque

class TaskScheduler:
    def __init__(self):
        self.queue = deque()

    def add_task(self, description, priority):
        self.queue.append((description, priority))

    def execute_next_task(self):
        if not self.queue:
            print("No tasks in the queue.")
            return

        # Trouver la tâche ayant la plus haute priorité
        highest_priority = min(task[1] for task in self.queue)

        for i, task in enumerate(self.queue):
            if task[1] == highest_priority:
                description, priority = task
                del self.queue[i]
                print(f"Executing task: {description}")
                return

    def show_queue(self):
        print("Task Queue:", list(self.queue))


# Test
scheduler = TaskScheduler()

scheduler.add_task("Clean room", 2)
scheduler.add_task("Do homework", 1)
scheduler.add_task("Wash dishes", 1)

scheduler.execute_next_task()  # Do homework
scheduler.show_queue()
```

```python
from collections import deque

def check_palindrome(text):
    # Nettoyage : suppression des espaces et conversion en minuscules
    cleaned = ''.join(char.lower() for char in text if char != ' ')

    queue = deque(cleaned)

    while len(queue) > 1:
        front = queue.popleft()
        back = queue.pop()

        if front != back:
            return False

    return True


# Tests
print(check_palindrome("Racecar"))          # True
print(check_palindrome("hello"))            # False
print(check_palindrome("A man a plan a canal Panama"))  # True
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_node = Node(value)

        if self.head is None:
            self.head = new_node
            return

        current = self.head

        while current.next:
            current = current.next

        current.next = new_node

    def display(self):
        current = self.head

        while current:
            print(current.value, end=" -> ")
            current = current.next

        print("None")
```

```python
def merge_sorted_linked_lists(list1, list2):
    dummy = Node(0)
    current = dummy

    p1 = list1.head
    p2 = list2.head

    while p1 and p2:
        if p1.value <= p2.value:
            current.next = p1
            p1 = p1.next
        else:
            current.next = p2
            p2 = p2.next

        current = current.next

    # Ajouter les éléments restants
    if p1:
        current.next = p1

    if p2:
        current.next = p2

    return dummy.next
```

```python
list1 = LinkedList()
list2 = LinkedList()

for value in [1, 3, 5]:
    list1.append(value)

for value in [2, 4, 6]:
    list2.append(value)

merged_head = merge_sorted_linked_lists(list1, list2)

current = merged_head

while current:
    print(current.value, end=" -> ")
    current = current.next

print("None")
```
