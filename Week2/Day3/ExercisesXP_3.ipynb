# Analyse Statistique avec SciPy
Ce notebook présente différentes méthodes d'analyses statistiques descriptives et inférentielles en Python.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from scipy.stats import linregress, norm, binom, pearsonr, spearmanr

print("Version de SciPy :", scipy.__version__)
```

## 1. Statistiques Descriptives

```python
data_desc = [12, 15, 13, 12, 18, 20, 22, 21]

mean_val = np.mean(data_desc)
median_val = np.median(data_desc)
variance_val = np.var(data_desc)
std_val = np.std(data_desc)
mode_val = stats.mode(data_desc, keepdims=True)

print("\n--- Statistiques descriptives ---")
print("Moyenne :", mean_val)
print("Médiane :", median_val)
print("Variance :", variance_val)
print("Écart-type :", std_val)
print("Mode :", mode_val.mode[0])
```

## 2. Distribution Normale

```python
mu, sigma = 50, 10

x = np.linspace(10, 90, 1000)
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Distribution Normale (μ=50, σ=10)")
plt.xlabel("Valeurs")
plt.ylabel("Densité")
plt.grid(True)
plt.show()
```

## 3. Test t de Student

```python
np.random.seed(42)

group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(60, 10, 100)

t_stat, p_val = stats.ttest_ind(group1, group2)

print("\n--- Test t ---")
print("T-statistic :", t_stat)
print("P-value :", p_val)

alpha = 0.05
if p_val < alpha:
    print("➡ Différence significative entre les groupes")
else:
    print("➡ Pas de différence significative")

plt.figure()
plt.hist(group1, bins=15, alpha=0.6, label="Groupe 1")
plt.hist(group2, bins=15, alpha=0.6, label="Groupe 2")
plt.legend()
plt.title("Comparaison des groupes")
plt.show()
```

## 4. Régression Linéaire

```python
sizes = np.array([50, 70, 80, 100, 120])
prices = np.array([150000, 200000, 210000, 250000, 280000])

result = linregress(sizes, prices)

slope = result.slope
intercept = result.intercept

print("\n--- Régression linéaire ---")
print("Slope :", slope)
print("Intercept :", intercept)

pred_size = 90
pred_price = slope * pred_size + intercept
print("Prix prédit pour 90 m² :", pred_price)

plt.figure()
plt.scatter(sizes, prices)
plt.plot(sizes, slope * sizes + intercept)
plt.title("Régression linéaire")
plt.xlabel("Taille")
plt.ylabel("Prix")
plt.show()
```

## 5. Analyse de la Variance (ANOVA)

```python
fert1 = [5, 6, 7, 6, 5]
fert2 = [7, 8, 7, 9, 8]
fert3 = [4, 5, 4, 3, 4]

f_val, p_val = stats.f_oneway(fert1, fert2, fert3)

print("\n--- ANOVA ---")
print("F-value :", f_val)
print("P-value :", p_val)

plt.figure()
plt.boxplot([fert1, fert2, fert3], labels warm=["F1", "F2", "F3"])
plt.title("ANOVA - Fertilisants")
plt.show()
```

## 6. Distribution Binomiale

```python
n, p, k = 10, 0.5, 5

prob = binom.pmf(k, n, p)

print("\n--- Binomiale ---")
print("P(X=5) :", prob)

x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

plt.figure()
plt.bar(x, y)
plt.title("Binomiale (n=10, p=0.5)")
plt.show()
```

## 7. Analyses de Corrélation

```python
df = pd.DataFrame({
    "age": [23, 25, 30, 35, 40],
    "income": [35000, 40000, 50000, 60000, 70000]
})

pearson_corr, pearson_p = pearsonr(df["age"], df["income"])
spearman_corr, spearman_p = spearmanr(df["age"], df["income"])

print("\n--- Corrélation ---")
print("Pearson :", pearson_corr)
print("Spearman :", spearman_corr)

plt.figure()
plt.scatter(df["age"], df["income"])
plt.title("Âge vs Revenu")
plt.show()
```
