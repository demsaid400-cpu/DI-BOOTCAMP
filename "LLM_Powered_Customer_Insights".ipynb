## 📊 Ce que le Projet Démontre

### 1. Appel API LLM Direct
```python
client = anthropic.Anthropic(api_key=API_KEY)
response = client.messages.create(
    model="claude-sonnet-4-6",
    max_tokens=512,
    temperature=0.1,
    system=SYSTEM_PROMPT,  # Prompt Engineering structuré JSON
    messages=[{"role": "user", "content": f"Analyse: {text}"}]
)
result = json.loads(response.content[0].text)
```

### 2. Prompt Engineering (System Prompt)
Retour JSON structuré avec 8 dimensions :
- `sentiment` · `confidence` · `topics` · `emotion`
- `key_aspects` (qualité, livraison, service, prix)
- `urgency` · `recommended_action`

### 3. Robustesse Production
- Retry logic (3 tentatives)
- Rate limiting (délai configurable)
- Gestion erreurs API + JSON parsing

### 4. Feature Engineering
Les features LLM enrichissent un RandomForest aval :
- Encodage One-Hot des variables catégorielles LLM
- Feature importance analysis
- Comparaison multi-approches

---

## 📈 Résultats

| Approche | Accuracy |
|---|---|
| TF-IDF + Logistic Regression | ~75% |
| LLM Claude Sonnet 4.6 (zero-shot) | ~82% |
| LLM Features + Random Forest | ~88% |

**Insight clé :** Le LLM n'est pas en compétition avec le ML classique — il l'**enrichit**.

## 🗺️ Architecture du Pipeline

```
Avis Client (texte brut)
        │
        ▼
┌───────────────────────┐
│   Prompt Engineering  │  → System prompt JSON strict
│   API Claude Sonnet   │  → Temperature 0.1, max_tokens 512
└───────────────────────┘
        │
        ▼
┌───────────────────────┐
│   JSON Parsing        │  → 8 dimensions structurées
│   Error Handling      │  → Retry + fallback
└───────────────────────┘
        │
        ├──────────────────────────────────────┐
        ▼                                      ▼
┌───────────────┐                    ┌──────────────────┐
│ LLM Direct    │                    │ Feature Encoding  │
│ Sentiment     │                    │ → Random Forest   │
│ Classification│                    │ (ML Downstream)   │
└───────────────┘                    └──────────────────┘
        │                                      │
        ▼                                      ▼
  api_results.json                    llm_features.csv
```
