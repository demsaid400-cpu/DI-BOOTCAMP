```python
import random

def jeu_de_devinette():
    print("🎯 Bienvenue dans le Défi de Devinette de Nombre ! 🎯")
    print("Je choisis un nombre entre 1 et 100.")
    print("Tu as 7 tentatives pour le trouver. Bonne chance ! 🚀\n")

    # Génération du nombre secret entre 1 et 100 inclus
    nombre_secret = random.randint(1, 100)
    tentatives_max = 7

    for essai in range(1, tentatives_max + 1):
        # Gestion des erreurs pour s'assurer que l'utilisateur entre un entier
        try:
            devinette = int(input(f"Tentative {essai}/{tentatives_max} - Entre ton chiffre : "))
        except ValueError:
            print("❌ Ce n'est pas un nombre valide. Tu perds une tentative !")
            continue

        # Vérification de la devinette
        if devinette == nombre_secret:
            print(f"🎉 Félicitations ! Tu as trouvé le nombre secret {nombre_secret} en {essai} tentatives ! 🏆")
            return
        elif devinette < nombre_secret:
            print("📉 Trop bas ! Essaie un nombre plus grand. 📈")
        else:
            print("📈 Trop haut ! Essaie un nombre plus petit. 📉")

    # Si le joueur épuise ses 7 tentatives sans trouver
    print(f"\n😢 Dommage ! Tu as épuisé tes 7 tentatives. Le nombre secret était {nombre_secret}. Entraîne-toi encore !")

# Lancement du jeu
if __name__ == "__main__":
    jeu_de_devinette()
```
